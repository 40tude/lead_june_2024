Télécharger et copier le répertoire d'installation de Flappy Birds

```python
import urllib.request
import zipfile
from shutil import copyfile, unpack_archive, copytree, rmtree
import os

# Define paths
PACKAGE_NAME = "flappy_bird_gym"
FLAPPY_BIRD_GITHUB = "https://github.com/Talendar/flappy-bird-gym/archive/refs/tags/v0.3.0.zip"
ZIPPED_FILENAME = "flappy-bird/flappy_zipped.zip"
UNZIPPED_FOLDER = "flappy-bird/flappy_unzipped"

# Download source code from Github
print("[INFO] : Downloading source code from Github !")
urllib.request.urlretrieve(FLAPPY_BIRD_GITHUB, ZIPPED_FILENAME)

# Unzip downloaded archive
print("[INFO] : Unzipping the archive !")
unpack_archive(ZIPPED_FILENAME, UNZIPPED_FOLDER)

# Copy relevant folder from unzipped folder
print("[INFO] : Copying the package code folder !")
try:
    copytree(os.path.join(UNZIPPED_FOLDER, "flappy-bird-gym-0.3.0", PACKAGE_NAME), os.path.join("flappy-bird", PACKAGE_NAME))
except FileExistsError:
    print("[INFO] : File already exists ! Skipping copy !")

# Delete unnecessary files
print("[INFO] : Cleaning unnecessary files !")
rmtree(UNZIPPED_FOLDER)
os.remove(ZIPPED_FILENAME)
print("[INFO] : All done !")
```

```python
%%writefile flappy-bird/config.py
import flappy_bird_gym
from ray.tune.registry import register_env
import gym

def env_creator(env_config):
    import flappy_bird_gym
    env = flappy_bird_gym.make("FlappyBird-v0")
    return env

register_env("flappy_bird", env_creator)

# Configure the algorithm.

config = {
    # Environment (RLlib understands openAI gym registered strings).
    "env": "flappy_bird",
    #"env": "FlappyBird-v0",


    # Use 2 environment workers (aka "rollout workers") that parallelly
    # collect samples from their own environment clone(s).
    "num_workers": 0,
    # Change this to "framework: torch", if you are using PyTorch.
    # Also, use "framework: tf2" for tf2.x eager execution.
    "framework": "tf",
    # Tweak the default model provided automatically by RLlib,
    # given the environment's observation- and action spaces.
    "render_env": False,

    "train_batch_size": 256,

    "preprocessor_pref": "deepmind",

    "model": {
        "fcnet_hiddens": [32, 16],
        "fcnet_activation": "relu",
    },
    "evaluation_interval":1,
    "exploration_config": {
        "type": "EpsilonGreedy",
        "initial_epsilon": 1.0,
        "final_epsilon": 0.1,
        "epsilon_timesteps": 100000, }
}
```

<h2 style="text-align: left; color:#20a08d; font-size: 20px"><span><strong>Entrainement de l'agent
</strong></span></h2>

- **Register external envs in Ray** : https://docs.ray.io/en/latest/rllib/rllib-env.html#configuring-environments
- **StackOverflow thread about registering envs** : https://stackoverflow.com/questions/58551029/rllib-use-custom-registered-environments

```python
%%writefile flappy-bird/train.py
# Import the RL algorithm (Trainer) we would like to use.

from ray.rllib.algorithms.ppo import PPO
import ray
from config import *

from ray.tune.logger import pretty_print

# Create our RLlib Trainer.
trainer = PPO(config=config)


# Run it for n training iterations. A training iteration includes
# parallel sample collection by the environment workers as well as
# loss calculation on the collected batch and a model update.


max_iter=1000
for i in range(1, max_iter+1):
    print(f"#### ITERATION {i} ####")
    result = trainer.train()
    print(pretty_print(result))

    print("episode_reward_max:",result['episode_reward_max'], 'episode_reward_min:', result["episode_reward_min"], 'episode_reward_mean:', result["episode_reward_mean"])
    if i % max_iter == 0:
       checkpoint = trainer.save(f'{os.getcwd()}/flappy-bird/checkpoints')
       print("checkpoint saved at", checkpoint)
```

<h2 style="text-align: left; color:#20a08d; font-size: 20px"><span><strong>Test de l'agent
</strong></span></h2>

```python
%%writefile flappy-bird/test.py
# Configure the algorithm.
config = {
    # Environment (RLlib understands openAI gym registered strings).
    "env": "FlappyBird-v0",

    # Use 2 environment workers (aka "rollout workers") that parallelly
    # collect samples from their own environment clone(s).
    "num_workers": 0,
    # Change this to "framework: torch", if you are using PyTorch.
    # Also, use "framework: tf2" for tf2.x eager execution.
    "framework": "tf",
    # Tweak the default model provided automatically by RLlib,
    # given the environment's observation- and action spaces.
    "render_env": False,

    "train_batch_size": 256,

    "preprocessor_pref": "Deepmind",

    "model": {
        "fcnet_hiddens": [32, 16],
        "fcnet_activation": "relu",
    },
}
# Create our RLlib Trainer.
trainer = PPOTrainer(config=config)

checkpoint_path="./checkpoints/checkpoint_001000/checkpoint-1000"
trainer.restore(checkpoint_path)

from ray.rllib.models.preprocessors import get_preprocessor

# Create the env to do inference in.
env = flappy_bird_gym.make("FlappyBird-v0")

prep = get_preprocessor(env.observation_space)(env.observation_space)

print("#### ENV PREPROCESSED SHAPE ####")
print(prep.transform(env.reset()).shape)
obs = prep.transform(env.reset())

episode_reward = 0.0

for _ in range(1000):

    print("### COMPUTING ACTION ####")
    # Compute an action (`a`).
    a = trainer.compute_single_action(obs)

    # Send the computed action `a` to the env.
    obs, reward, done, _ = env.step(a)
    episode_reward += reward
    # Is the episode `done`? -> Reset.
    env.render()
    time.sleep(1/30)

    if done:
        print(f"Episode done: Total reward = {episode_reward}")
        env.reset()
```
