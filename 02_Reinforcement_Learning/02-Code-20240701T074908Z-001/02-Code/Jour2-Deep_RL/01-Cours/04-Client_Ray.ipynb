<h2 style="text-align: left; color:#20a08d; font-size: 25px"><span><strong>Installation
</strong></span></h2>

Avant de commencer ce cours, vous devrez :

- Avoir un cluster Ray opérationnel (soit localement avec minikube, soit à distance)
- Consulter un cours sur le machine learning distribué
- Vous devrez d'abord cloner le dépôt git officiel sur votre ordinateur local :
    - <code>git clone https://github.com/ray-project/kuberay.git</code>
- Démarrer minikube
    - <code>minikube start --driver=docker</code>

- <code>cd</code> dans le répertoire ray : <code>cd kuberay/helm-chart/ray-cluster/</code>
- Si vous avez déjà votre service en cours d'exécution, vous devriez le voir avec <code>helm list</code>
    - Installez le cluster Ray à l'aide de Helm :<code>helm -n [NAMESPACE] install [RELEASE_NAME] --create-namespace ./ray</code>
    - Remplacez <code>[RELEASE_NAME]</code> par un nom qui vous convient.
    Avec la commande ci-dessus, Helm créera un nouvel espace de noms. Si vous ne le souhaitez pas, supprimez simplement <code>--create-namespace </code> et <code>-n [NAMESPACE]</code>.

<h2 style="text-align: left; color:#20a08d; font-size: 25px"><span><strong>Qu'est-ce que RayClient ?

</strong></span></h2>

Ray Client est un excellent outil pour exploiter de manière interactive un cluster Ray distant.

Fondamentalement, vous écrirez du code python et l'exécuterez directement par votre cluster (au lieu d'empaqueter tout le code et de l'envoyer à votre cluster comme vous le feriez avec Ray Job Submission).

````python
!pip3 install -r requirements.txt --force-reinstall
````

```python
import ray
from config import config

ray.init(
    "ray://127.0.0.1:10001",
    runtime_env={"working_dir": ".", "pip": "./requirements.txt"},
    )

from ray.rllib.agents.ppo import PPOTrainer


# Create our RLlib Trainer.
trainer = PPOTrainer(config=config)

# Run it for n training iterations. A training iteration includes
# parallel sample collection by the environment workers as well as
# loss calculation on the collected batch and a model update.
max_range=1000
for i in range(1, max_range+1):
    print(f"#### ITERATION {i} ####")
    result = trainer.train()
    if i % max_range == 0:
       checkpoint = trainer.save()
       print("checkpoint saved at", checkpoint)
```

````python
%%writefile ray_cluster/train.py
import ray
from config import config

ray.init(
    "ray://127.0.0.1:10001",
    runtime_env={"working_dir": ".", "pip": "./requirements.txt"},
    )

from ray.rllib.agents.ppo import PPOTrainer


# Create our RLlib Trainer.
trainer = PPOTrainer(config=config)

# Run it for n training iterations. A training iteration includes
# parallel sample collection by the environment workers as well as
# loss calculation on the collected batch and a model update.
max_range=1000
for i in range(1, max_range+1):
    print(f"#### ITERATION {i} ####")
    result = trainer.train()
    if i % max_range == 0:
       checkpoint = trainer.save()
       print("checkpoint saved at", checkpoint)
````

<h3 style="text-align: left; color:#20a08d; font-size: 20px"><span><strong>Contraintes
</strong></span></h3>

Avant d'utiliser Ray Client (que vous pourriez trouver si magique 🪄), vous devez être conscient que **vous devez avoir une version de python correspondante entre votre machine locale et votre cluster** , c'est-à-dire que si votre cluster ray a python==3.7 installé, vous devez avoir python==3.7 sur votre machine locale.

C'est pourquoi, vous devez utiliser l'environnement conda ou pip. De cette façon, vous pouvez configurer votre environnement conda de cette façon :

```bash
conda create -n rllib python=3.7.7
conda activate rllib
pip install -r requirements.txt
```
